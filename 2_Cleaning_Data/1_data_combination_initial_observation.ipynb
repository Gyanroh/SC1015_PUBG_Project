### Killdata


```python
killpath = r'C:/Users/swrgo/Desktop/aNTU STUFF/DataCleaning/KillData' # @fraser @kai, use your individual PC path here
genpath  = r'C:/Users/swrgo/Desktop/aNTU STUFF/DataCleaning/GenData'   #PC path for gendata
```

```python
import os
import glob

allKillFiles = glob.glob(os.path.join(killpath , "*.csv"))
killarray = []

for filename in allKillFiles:
    print("imported ",filename)
    killdf = pd.read_csv(filename, index_col=None, header=0)
    killarray.append(killdf)

raw_kill = pd.concat(killarray, axis=0, ignore_index=True)
```

### General Data

```python
allGenFiles = glob.glob(os.path.join(genpath , "*.csv"))
genarray = []

for filename in allGenFiles:
    print("imported ",filename)
    gendf = pd.read_csv(filename, index_col=None, header=0)
    genarray.append(gendf)

raw_gen = pd.concat(genarray, axis=0, ignore_index=True)
raw_gen
```


## Basic Cleaning

Primary focus on most popular map, Erangel

```python
raw_kill.columns
```

```python
#Select only Erangel for kill_data
raw_kill.loc[raw_kill['map'].isin(["Baltic_Main", "Erangel_Main"])]
erangel_kill = raw_kill.drop('map', axis = 1)
erangel_kill
```

We can observe that some of the data is suspiciously invalid. For example, the killer_position cannot be at (0, 0) for a non-bluezone death, as this implies that the person who slayed the victim is at an invalid coordinate. We shall drop it. 

```python
erangel_kill = erangel_kill[erangel_kill['killer_position_x'] !=0]  #drop (0, 0)
erangel_kill = erangel_kill[erangel_kill['victim_position_x'] !=0]
```

```python
pd.DataFrame(erangel_kill).to_csv('kill_match_stats.csv', index=False)  # approx 300 MB
```
